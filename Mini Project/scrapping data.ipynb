```python
import pandas as pd
from google_play_scraper import Sort, reviews

apps = {
    "gopay": "com.gojek.app",
    "ovo": "ovo.id",
    "dana": "id.dana",
    "shopeepay": "com.shopee.id"
}

TARGET_REVIEW = 1250

for app_name, package_name in apps.items():
    print(f"Scraping {app_name}...")

    all_reviews = []
    continuation_token = None

    while len(all_reviews) < TARGET_REVIEW:
        result, continuation_token = reviews(
            package_name,
            lang='id',
            country='id',
            sort=Sort.MOST_RELEVANT,
            count=200,
            filter_score_with=None,
            continuation_token=continuation_token
        )

        if not result:
            break

        all_reviews.extend(result)

        if continuation_token is None:
            break

    # Potong tepat 1.250 data
    all_reviews = all_reviews[:TARGET_REVIEW]

    df = pd.DataFrame(all_reviews)
    df = df[['content']]

    file_name = f"review_{app_name}.csv"
    df.to_csv(file_name, index=False, encoding='utf-8-sig')

    print(f"{app_name}: {len(df)} data disimpan ke {file_name}")
```

```python
import pandas as pd

# Load masing-masing CSV
df_gopay = pd.read_csv("review_gopay.csv")
df_ovo = pd.read_csv("review_ovo.csv")
df_dana = pd.read_csv("review_dana.csv")
df_shopeepay = pd.read_csv("review_shopeepay.csv")

# Tambah kolom app
df_gopay['app'] = 'gopay'
df_ovo['app'] = 'ovo'
df_dana['app'] = 'dana'
df_shopeepay['app'] = 'shopeepay'

# Gabungkan semua data
df_all = pd.concat(
    [df_gopay, df_ovo, df_dana, df_shopeepay],
    ignore_index=True
)

# Cek jumlah data
print("Total data:", df_all.shape[0])
print(df_all.head())

# Simpan ke satu CSV
df_all.to_csv("ewallet_all.csv", index=False, encoding='utf-8-sig')
```
